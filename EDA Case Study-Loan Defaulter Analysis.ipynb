# Data Cleaning

Checking for missing values

```python
#Calculating the Percentage of missing values in the columns
prev_missingdata = (prev_ap.isnull().sum() / len(prev_ap)) * 100
prev_missingdata.sort_values(ascending = False, inplace = True)
prev_missingdata
```

#### Dropping unwanted columns

```python
prev_ap.drop(['RATE_INTEREST_PRIMARY','RATE_INTEREST_PRIVILEGED'], axis=1, inplace=True)
prev_ap.dropna(subset=['AMT_CREDIT','PRODUCT_COMBINATION'], inplace=True)

#Checking the remaining columns
prev_ap
```

```python
print(prev_ap['NAME_TYPE_SUITE'].value_counts())
```

```python
# Filling missing value as 'Unaccompanied' as most common value

prev_ap['NAME_TYPE_SUITE'].fillna(prev_ap.NAME_TYPE_SUITE.mode()[0], inplace=True)
```

```python
num_feat = []
for col in prev_ap.columns:
    if prev_ap[col].dtypes==float or prev_ap[col].dtypes==int:
        num_feat.append(col)
            
print (num_feat)
```

```python
prev_num_df = pd.DataFrame()

for col in num_feat:
    prev_num_df[col] = prev_ap[col]
    
prev_num_df.info()
```

```python
# Checking correlation between numeric features of previous application data

plt.figure(figsize=(20,12))
sns.heatmap(prev_num_df.corr(), annot=True)
plt.show()
```

```python
# This gives percentage of defaulter in total applicants having the same label of the category.
# For example: In 100 applicants, whose'NAME_GOODS_CATEGORY' is Insurance
# approx. 10 people have payment difficulty

def value_wise_defaulter_percentage(df, col):
    new_df = pd.DataFrame(columns=['Value', 'Percentage of Defaulter'])
    
    for value in df[col].unique():
        default_cnt = df[(df[col] == value) & (df.TARGET == 1)].shape[0]
        total_cnt = df[df[col] == value].shape[0]
        new_df = new_df.append({'Value' : value , 'Percentage of Defaulter' : (default_cnt*100/total_cnt)}, ignore_index=True)
    return new_df.sort_values(by='Percentage of Defaulter', ascending=False)
```

```python
value_wise_defaulter_percentage(prev_merged, 'NAME_GOODS_CATEGORY')
```

Observation:

Applicants who applied for Insurance and Vehicles indicates highest percentage of default cases in previous application data set 

```python
value_wise_defaulter_percentage(prev_merged, 'NAME_PORTFOLIO')
```

Percentage of Default case is high in 'Cards' mode.
